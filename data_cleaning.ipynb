```python
import pandas as pd
```

Load the data and combine it into one dataframe.

```python
root_folder = "./Data"
jun24 = pd.read_csv(f'{root_folder}/202406-divvy-tripdata.csv', index_col=0)
jul24 = pd.read_csv(f'{root_folder}/202407-divvy-tripdata.csv', index_col=0)
aug24 = pd.read_csv(f'{root_folder}/202408-divvy-tripdata.csv', index_col=0)
sept24 = pd.read_csv(f'{root_folder}/202409-divvy-tripdata.csv', index_col=0)
oct24 = pd.read_csv(f'{root_folder}/202410-divvy-tripdata.csv', index_col=0)
nov24 = pd.read_csv(f'{root_folder}/202411-divvy-tripdata.csv', index_col=0)
dec24 = pd.read_csv(f'{root_folder}/202412-divvy-tripdata.csv', index_col=0)
jan25 = pd.read_csv(f'{root_folder}/202501-divvy-tripdata.csv', index_col=0)
feb25 = pd.read_csv(f'{root_folder}/202502-divvy-tripdata.csv', index_col=0)
mar25 = pd.read_csv(f'{root_folder}/202503-divvy-tripdata.csv', index_col=0)
apr25 = pd.read_csv(f'{root_folder}/202504-divvy-tripdata.csv', index_col=0)
may25 = pd.read_csv(f'{root_folder}/202505-divvy-tripdata.csv', index_col=0)
```

```python
data = pd.concat([jun24, jul24, aug24, sept24, oct24, nov24, dec24, jan25, feb25, mar25, apr25, may25])
data = data.reset_index(level=0)
data.info()
```

Change the datatype of datetime columns

```python
data['started_at'] = pd.to_datetime(data['started_at'])
data['ended_at'] = pd.to_datetime(data['ended_at'])
data.info()
```

Dropping all the rows with null values

```python
data_no_null = data.dropna()
data_no_null.shape
```

Creating ride time column in seconds

```python
df_ride = data_no_null.copy(deep=True)

df_ride['ride_time[s]'] = (df_ride['ended_at'] - df_ride['started_at']).dt.total_seconds()

df_ride.sample(10)
```

Checking if there are rides with ride time less than 0. It may be because of error in sourcing the data or other application glitch

```python
df_ride[df_ride['ride_time[s]'] < 0]
```

```python
df_ride[df_ride['ride_time[s]'] < 0].shape
```

There are 26 records with negative ride time. I'll delete the rows with negative ride time.

```python
df_ride = df_ride[df_ride['ride_time[s]'] > 0]
```

Check if there are records with same starting and ending stations

```python
df_same_se = df_ride[df_ride['start_station_id'] == df_ride['end_station_id']]
df_same_se.info()
```

Rides with ride time of less than 60secs and having the same starting and ending station, can be incorrect logging/testing/completely irrelevant

```python
df_same_se[df_same_se['ride_time[s]']<60]
```

23037 out of 198846 (11.6%) of the rides with same source and destination have ride time of less than 60secs

Removing such rides from the final data

```python
df_clean = df_ride[~((df_ride['start_station_id'] == df_ride['end_station_id']) & (df_ride['ride_time[s]'] < 60))]
df_clean.shape
```

```python
df_final = df_clean.copy(deep=True)
df_final['Weekday'] = df_final['started_at'].dt.weekday
```

Renaming the columns for analysis

```python
new_columns = {
    'rideable_type':'BikeType',
    'started_at': 'RideStart',
    'ended_at': 'RideEnd',
    'ride_time[s]': 'RideTime[s]',
    'start_station_name': 'StartStation',
    'end_station_name': 'EndStation',
    'member_casual': 'UserType'
}

df_final.rename(columns=new_columns, inplace=True)
```
